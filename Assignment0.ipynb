```python
#Stage1 - Registration

import re
regex = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'
regex1 = r'^(?=.*[a-z])(?=.*[A-Z])(?=.*\d)(?=.*[@$!%*#?&])[A-Za-z\d@$!#%*?&]{5,16}$'
def check(email, pwd):
  if (re.fullmatch(regex, email)):
      print("valid email")
      for i in range(0,3):
        if (re.fullmatch(regex1, pwd)):
          print("valid password")
          break
        else:
          print("invalid password")
          pwd = input("Enter a password : ")         
  else:
    print("invalid email/username")
    email = input("Enter an Email : ")
    check(email, pwd)

if __name__ == '__main__':
  email = input("Enter an Email : ")
  print('Pwd contains 1 capital letter, 1 number and 1 special character & length of the pwd between 5 and 16')
  password = input("Enter a password : ")
  check(email, password)
  l = [email,password]
# print(l)
#Stage2, Stage3
with open("/content/sample.txt.txt","r") as f:
  data = f.readlines()
  count=False
  for line in data:
    word = line.split(",")
#    print(word)
    if (word==l):
      print('User present in the file')
      count=True
      break
    else: count=False
  print(count)
  if (count==False):
    with open("/content/sample.txt.txt","a") as f:
      f.write("\n")
      f.write(email)
      f.write(",")
      f.write(password)
      f.close()
  else: f.close()
```

```python
def countRotation(arr,y):
    min = arr[0]
    for i in range(0, y):
        if(min>arr[i]):
            min = arr[i]
            min_index = i
            return min_index
arr = input()
arr1 = input()
arr2 = input()
y = len(arr2)
if(len(arr1)==len(arr2)):
    print(countRotation(arr,y))
```

```python
def reverseWord(s):
    words = s.split(" ")
    newwords = [word[::-1] for word in words]
    newsentance = " ".join(newwords)
    return newsentance

x = input()
n = len(x)
if(n>=1 and n<=100000):
    print(reverseWord(x))
```
