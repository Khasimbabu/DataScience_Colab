```python
def read_file(filename):
  with open(filename,'r') as file:
    text = file.read()
  return text
text = read_file('/content/sample_data/shakespeare-taming.txt')
text[:100]
```

```python
nltk.download('punkt')
```

```python
# Tokenize words
words = word_tokenize(text)
print('Size as a list : ',len(words))
print('Size as a set : ', len(set(words)))
print(words[:100])
```

```python
# # Tokenize sentences
sentences = sent_tokenize(text)
print('Number of sentences : ',len(sentences),'\n')

for sentence in sentences[:5]:
  print(sentence.strip())
print('Size as a list : ',len(sentences))
print('Size as a set : ',len(set(sentences)),end='\n\n')
print(sentences[:10])
```

```python
# Regular Expressions
import re
if((re.search('^a','abc'))):
  print('Found it !')
if((re.search('^a','Abc'))):
  print('Found it !')
else:
  print('Not Found')

mywords = words[:100]

print(len(set([w for w in mywords if re.search('^a*',w)])))
print(set([w for w in words if re.search('^a*',w)]))  

print(len(set([w for w in mywords if re.search('^a+',w)])))
print(set([w for w in mywords if re.search('^a+',w)]))

print([w for w in mywords if re.search('^at+',w)])
print([w for w in mywords if re.search('^at*',w)])
print([w for w in mywords if re.search('^a*t*',w)])
print([w for w in mywords if re.search('^a+t+',w)])
print([w for w in mywords if re.search('^a*t+',w)])
```

```python
## Build a small application which will take input a name of a person and 
# tell/predict if the name is female or male
# choosing features which our algorithm will use to identify 
# whether the person is male or female
```

```python
def gender_features_part1(name):
  name = str(name).lower()
  return {'last_letter': name[-1:]} # returning a dictionary
print(gender_features_part1('Sam'))
print(gender_features_part1('Sowmya'))
```

```python
import nltk
from nltk.corpus import gutenberg as gt
from nltk.corpus import PlaintextCorpusReader
from nltk.tokenize import word_tokenize,sent_tokenize
from nltk import PorterStemmer,LancasterStemmer
from nltk.corpus import names as names_sample
import random
```

```python
nltk.download('names')
```

```python
names = [(name,'male') for name in names_sample.words('male.txt')] + [(name,'female') for name in names_sample.words('female.txt')]
print(names)
```

```python
random.shuffle(names)
```

```python
for name,gender in names[:20]:
  print('Name:', name, '  Gender:',gender)
```

```python
# make a feature set for all the names
feature_sets = [(gender_features_part1(name.lower()),gender) for name,gender in names]
feature_sets
```

```python
for dict,gender in feature_sets[:20]:
  print(dict,gender)  
```
