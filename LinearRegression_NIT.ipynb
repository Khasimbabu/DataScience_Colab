```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/sample_data/student_info.csv')
df.head()
```

```python
df.info()
```

```python
plt.scatter(x=df.study_hours,y=df.student_marks)
plt.xlabel("Students Study Hours")
plt.ylabel("Students Marks")
plt.title("Scatter Plot of Students Study Hours vs Students Marks")
plt.show()
```

```python
df.isna().sum()
```

```python
df.isnull().sum()
```

```python
df.mean()
```

```python
df1 = df.fillna(df.mean())
df1.isnull().sum()
```

```python
X = df1.drop("student_marks",axis="columns")
X
```

```python
y=df1.drop("study_hours",axis="columns")
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
print(model.coef_)
print(model.intercept_)
```

```python
print(model.predict([[4]])[0][0].round(2))
```

```python
print(model.predict(X_test))
```

```python
y_pred = model.predict(X_test)
df_new = pd.DataFrame(np.c_[X_test,y_test,y_pred],columns=["Study_Hours","Student_Marks","Predicted_Marks"])
df_new
```

```python
print(model.score(X_train,y_train))
print(model.score(X_test,y_test))
```

```python
print(plt.scatter(X_train,y_train))
print(plt.scatter(X_test,y_test))
plt.plot(X_train,model.predict(X_train),color='r')
```

```python
import joblib
joblib.dump(model, "student_mark_predictor.pkl")
```

```python
dataset_values = np.array(dataset) #converting list into an array
print(dataset_values)
```

```python
labelencoder_obj = LabelEncoder() 
integer_encode =labelencoder_obj.fit_transform(dataset_values) # converting the text data into numeric
print(integer_encode)
```

```python
integer_encode = integer_encode.reshape(len(integer_encode),1)
print(integer_encode)
```

```python
onehot_obj = OneHotEncoder(sparse=False)
onehot_encode = onehot_obj.fit_transform(integer_encode)  # converting to Binary format
print(onehot_encode)
```
