```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
df1 = df
```

```python
df1.fillna(df1.mean(),inplace=True)
```

```python
df1_cnt = np.array(df1['Country'])
```

```python
df1_cnt
```

```python
labelencoder_obj = LabelEncoder()
integerencode = labelencoder_obj.fit_transform(df1_cnt)
# Convert the text into numeric
print(integerencode)
```

```python
integerencode = integerencode.reshape(len(integerencode),1)
print(integerencode)
```

```python
onehotencoder_obj = OneHotEncoder(sparse=False)
onehotencode = onehotencoder_obj.fit_transform(integerencode)
print(onehotencode)
```

**Step 5: Creating a dummy variable**

```python
df1['CountryNew'] = df1['Country']
```

```python
df1['CountryNew'] = integerencode
```

```python
X = df1[['Purchased','Age','Salary','Country','CountryNew']]
```

**Step 6: Splitting the datasets into training sets and Test sets**

```python
X = df1[['Age','Salary','CountryNew']]
X
```

```python
y = df1['Purchased']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
pd.get_dummies(df1.Country)
```

**Step 7: Feature Scaling**

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
from sklearn.metrics import r2_score
print(r2_score(y_test,y_pred))  
print(r2_score(y_train,model.predict(X_train)))                     
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))  
print(accuracy_score(y_train,model.predict(X_train))) 
```
